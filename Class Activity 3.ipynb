```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')
plt.show
```

```python
categories = ['A','B','C']
values = [10,5,7]
plt.bar(categories,values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
x,y = data["Density\n(P/Km2)"],data["Country"]
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#x,y = data["Density\n(P/Km2)"],data["Country"]
x = [1,2,3,5,5]
y = [5,9,3,6,1]
plt.plot(x,y,linestyle='--',marker='o',color='red',label='Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Graph')
plt.legend()
plt.show
```

```python
categories = ['A','B','C']
values1 = [10,5,7]
values2 = [6,8,4]

plt.bar(categories,values1,label='Values 1')
plt.bar(categories,values2,label='Values 2',bottom=values1)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
np.random.seed(42) # Due to the seed value the distribution remains same all the time
data = np.random.randn(1000) #Returns a normal distribution
plt.hist(data,bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
```

```python
data = [10,15,20,25,30]
mean_value = np.mean(data)
median_value = np.median(data)
print("The mean is ",mean_value)
print("The median is ",median_value)
```

```python
std_dev = np.std(data)
print("Standard Deviation is ",std_dev)
```

```python
sample_data = [23,27,25,24,22,26]
population_mean = 25
#t_stat compares means of two groups
#p_value measures probability of values between two groups are due to chance
t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)
print("T-statistic: ", t_stat)
print("P-value: ", p_value)
```

```python
if p_value < 0.05:
    print("The sample mean is significantly different from the population mean")
else:
    print("There is no significant difference between the sample mean and the population mean")
```

```python
import matplotlib.pyplot as plt
```

```python
x =  np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
correlation_coef = np.corrcoef(x,y)[0,1]
print("Correlation Coefficient: ",correlation_coef)
```

```python
#data_new = np.array([[1,2,3,4,5],[2,4,6,8,10]])
data_new = np.array([[1,2],[2,4],[3,6],[4,8],[5,10]]) # (5,2) 5 rows and 2 columns
plt.imshow(data_new,cmap='hot')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
predicted_y = slope*x + intercept
plt.scatter(x,y,label='Data')
plt.plot(x,predicted_y,color='gold',label='Regression Line')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
